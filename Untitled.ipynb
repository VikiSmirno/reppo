```python
df['MedHouseVal'] = data['target'] # Медианная стоимость дома
```

```python
# 2. Разделение данных
X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,  # 20% данных для теста (стандартный выбор)
    random_state=42
)
```

```python
def calculate_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    
    return {
        'MSE': mse,
        'R2': r2,
        'MAPE': mape
    }
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import optuna
```

```python
# 3. Функция для Optuna
def objective(trial):
    # Диапазоны гиперпараметров
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 1),
        'reg_lambda': trial.suggest_float('reg_lambda', 1, 2),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'random_state': 42
    }
    
    # Создание и обучение модели
    model = xgb.XGBRegressor(**params)
    model.fit(X_train_scaled, y_train)
    
    # Предсказание и оценка
    y_pred = model.predict(X_test_scaled)
    mse = mean_squared_error(y_test, y_pred)
    
    return mse
```

```python
# 4. Подбор гиперпараметров с помощью Optuna
print("\n=== Подбор гиперпараметров с помощью Optuna ===")
study = optuna.create_study(
    direction='minimize'
)

# Запуск оптимизации
study.optimize(objective, n_trials=50, show_progress_bar=True)
```

```python
print(f"\nЛучшие гиперпараметры:")
for key, value in study.best_params.items():
    print(f"{key}: {value}")

print(f"Лучшее MSE: {study.best_value:.4f}")
```

```python
print("\n=== Обучение финальной модели ===")
best_params = study.best_params
best_params['random_state'] = 42

final_model = xgb.XGBRegressor(**best_params)
final_model.fit(X_train_scaled, y_train)

# 6. Предсказание и оценка качества
y_pred_train = final_model.predict(X_train_scaled)
y_pred_test = final_model.predict(X_test_scaled)

print("\n=== Результаты на тренировочной выборке ===")
train_metrics = calculate_metrics(y_train, y_pred_train)
for metric, value in train_metrics.items():
    print(f"{metric}: {value:.4f}")

print("\n=== Результаты на тестовой выборке ===")
test_metrics = calculate_metrics(y_test, y_pred_test)
for metric, value in test_metrics.items():
    print(f"{metric}: {value:.4f}")
```

# ДОПОЛНИТЕЛЬНАЯ ЧАСТЬ

```python
def gen_objective(estimator_class, grid, X_train, y_train, X_test, y_test, **kwargs):
    def objective(trial):
        params = {}
        for param_name, param_config in grid.items():
            if isinstance(param_config, dict):
                param_type = param_config['type']
                if param_type == 'int':
                    params[param_name] = trial.suggest_int(
                        param_name, 
                        param_config['low'], 
                        param_config['high']
                    )
                elif param_type == 'float':
                    params[param_name] = trial.suggest_float(
                        param_name, 
                        param_config['low'], 
                        param_config['high'],
                        log=param_config.get('log', False)
                    )
                elif param_type == 'categorical':
                    params[param_name] = trial.suggest_categorical(
                        param_name, 
                        param_config['choices']
                    )
            else:
                params[param_name] = param_config
        
        params.update(kwargs)
        
        # Создаем и обучаем модель
        model = estimator_class(**params)
        model.fit(X_train, y_train)
        
        # Предсказание и оценка
        y_pred = model.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        
        return mse
    
    return objective
```

```python
# Примеры сеток гиперпараметров для разных моделей
xgb_grid = {
    'n_estimators': {'type': 'int', 'low': 100, 'high': 1000}, #количество деревьев
    'max_depth': {'type': 'int', 'low': 3, 'high': 10}, # максимальная глубина
    'learning_rate': {'type': 'float', 'low': 0.01, 'high': 0.3, 'log': True}, #скорость обучения
    'subsample': {'type': 'float', 'low': 0.6, 'high': 1.0}, # доля случайных объектов 
    'colsample_bytree': {'type': 'float', 'low': 0.6, 'high': 1.0}, # доля случайных признаков 
    'reg_alpha': {'type': 'float', 'low': 0, 'high': 1}, # L1 регуляризация
    'reg_lambda': {'type': 'float', 'low': 1, 'high': 2}, # L2 регуляризация
    'min_child_weight': {'type': 'int', 'low': 1, 'high': 10}, # минимальная сумма весов (или количество объектов) в листе
}

lgb_grid = {
    'n_estimators': {'type': 'int', 'low': 100, 'high': 1000},
    'max_depth': {'type': 'int', 'low': 3, 'high': 10},
    'learning_rate': {'type': 'float', 'low': 0.01, 'high': 0.3, 'log': True},
    'subsample': {'type': 'float', 'low': 0.6, 'high': 1.0},
    'colsample_bytree': {'type': 'float', 'low': 0.6, 'high': 1.0},
    'reg_alpha': {'type': 'float', 'low': 0, 'high': 1},
    'reg_lambda': {'type': 'float', 'low': 1, 'high': 2},
    'min_child_weight': {'type': 'int', 'low': 1, 'high': 10},
    'num_leaves': {'type': 'int', 'low': 20, 'high': 100}, # максимальное количество листьев
}
```

```python
def compare_models(models_config, X_train, y_train, X_test, y_test, n_trials=50):
    results = {}
    
    for model_name, config in models_config.items():
        print(f"Оптимизация гиперпараметров для {model_name}...")
        
        objective_func = gen_objective(
            estimator_class=config['class'],
            grid=config['grid'],
            X_train=X_train,
            y_train=y_train,
            X_test=X_test,
            y_test=y_test,
            random_state=42,
            **config.get('extra_params', {})
        )
        
        study = optuna.create_study(direction='minimize')
        study.optimize(objective_func, n_trials=n_trials)
        
        results[model_name] = {
            'best_params': study.best_params,
            'best_value': study.best_value,
            'study': study
        }
        
        print(f"Лучшее MSE для {model_name}: {study.best_value:.4f}")
        print(f"Лучшие параметры: {study.best_params}")
        print("-" * 50)
    
    return results
```
